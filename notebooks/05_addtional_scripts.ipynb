# ADDITIONAL SCRIPTS TO GENERATE PLOTS

```python
import matplotlib.pyplot as plt
import numpy as np

# F1-scores from your reports
models = ["Logistic Regression", "Random Forest", "Neural Network"]
f1_unbalanced = [0.27, 0.31, 0.37]
f1_balanced   = [0.27, 0.32, 0.44]

x = np.arange(len(models))
width = 0.35

plt.figure(figsize=(9,5))
plt.bar(x - width/2, f1_unbalanced, width, label="Unbalanced", color="#1f77b4")
plt.bar(x + width/2, f1_balanced, width, label="Balanced", color="#ff7f0e")

plt.ylabel("F1-score")
plt.title("Comparison of F1-scores Across Models (Balanced vs Unbalanced)")
plt.xticks(x, models, rotation=15)
plt.ylim(0, 0.60)
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.legend()
plt.tight_layout()

plt.savefig("res/f1_comparison.png", dpi=300)
plt.show()
```
